```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto = pd.read_csv("titanic.csv")
auto.head()
```

```python
import matplotlib.pyplot as plt

# Create a histogram of ages
plt.figure(figsize=(10, 6))
plt.hist(auto['age'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python
# Create a bar plot for the number of passengers in each class
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=auto)
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Create a scatter plot comparing Age and Fare
plt.figure(figsize=(10, 6))
plt.bar(auto['age'], auto['fare'], alpha=0.5)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```
