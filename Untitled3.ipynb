```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks github.csv')
sm
```

```python
sm['Total']=sm[['English', 'Hindi', 'Maths', 'Science', 'SST']].sum(axis=1)
sm
```

```python
sm['Percentage']=sm['Total']/5
sm
```

```python
def clac_grade(p):
    if p<= 100 and p>= 80:
        return 'A'
    elif p<80 and p>=60:
        return 'B'
    elif p<60 and p>=50:
        return 'C'
    elif p<50 and p>=40:
        return 'D'
    elif p<40 and p>=30:
        return 'E'
    elif p<30 and p>=0:
        return 'F'
    else:
        return'X'

sm['Grade']=sm['Percentage'].apply(clac_grade)
sm
```

```python
plt.figure(figsize=(10,6))
for sub in [ 'English', 'Hindi', 'Maths', 'Science', 'SST' ]:
      plt.plot(sm['Name'], sm[sub],marker='o', label=sub)
      plt.title("Marks in different Subjects")
      plt.xlabel('Name')
      plt.ylabel('Marks')
      plt.legend()
      plt.grid(True)
      plt.show()
```
